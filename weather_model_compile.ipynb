```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
nyc_data = pd.read_csv("C:/Users/Shakir/Downloads/nyc.txt")
nyc_data.drop(columns = nyc_data.columns[0], axis = 1, inplace= True)
nyc_data.Events.fillna("Normal", inplace=True)
```

```python
nyc_data = nyc_data.drop(columns=['city', 'season', 'WindDirDegrees.br...'])
#useless data
```

```python
nyc_data.PrecipitationIn.fillna(0, inplace=True)
nyc_data.PrecipitationIn.replace('T', '0.0', inplace=True)
```

```python
nyc_data['PrecipitationIn'] = pd.to_numeric(nyc_data['PrecipitationIn'])
```

```python
events = nyc_data.pop('Events')
dates = nyc_data.pop('Date')
```

```python
events.unique()
```

```python
Fog = []
Normal = []
Rain = []
Thunder = []
Snow = []
Unusual = []
```

```python
for i in events:
    if "Fog" in i:
        Fog.append(1)
    else:
        Fog.append(0)
    if "Rain" in i:
        Rain.append(1)
    else:
        Rain.append(0)
    if "Thunderstorm" in i:
        Thunder.append(1)
    else:
        Thunder.append(0)
    if "Snow" in i:
        Snow.append(1)
    else:
        Snow.append(0)
    if "Normal" in i:
        Normal.append(1)
    else:
        Normal.append(0)
    if "Tornado" in i or "Hail" in i:
        Unusual.append(1)
    else:
        Unusual.append(0)
```

```python
d = {'Normal' : Normal,  'Rain' : Rain, 'Thunder' : Thunder, 'Fog' : Fog, 'Snow': Snow, 'Unusual' : Unusual }
weather = pd.DataFrame(d)
weather
```

```python
nyc_data = pd.concat([nyc_data, weather], axis=1)
nyc_data.dtypes
```

```python
corr = nyc_data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
nyc_data = nyc_data.fillna(0)
nyc_data.isnull().any()
#replace nulls with 0
```

```python
nyc_data.columns
```

```python
inputdata = ['Max.TemperatureF', 'Mean.TemperatureF', 'Min.TemperatureF',
       'Max.Dew.PointF', 'MeanDew.PointF', 'Min.DewpointF', 'Max.Humidity',
       'Mean.Humidity', 'Min.Humidity', 'Max.Sea.Level.PressureIn',
       'Mean.Sea.Level.PressureIn', 'Min.Sea.Level.PressureIn']
outputdata = ['Normal', 'Rain', 'Thunder', 'Fog', 'Snow']
```

```python
inputTF = tf.convert_to_tensor(nyc_data[inputdata])
outputTF = tf.convert_to_tensor(nyc_data[outputdata])
#converting data into proper implementation for the machine model
```

```python
inputs = keras.Input(shape=(inputTF.shape[1],))
x = keras.layers.Dense(360, activation = "relu")(inputs)
x = keras.layers.Dense(180, activation = "sigmoid")(x)
#x = keras.layers.AlphaDropout(0.1)(x)
outputs = keras.layers.Dense(5, activation = "sigmoid")(x)
model = keras.Model(inputs,outputs)
model.summary()

#creating the machine learning model
```

```python
model.compile(optimizer='Nadam', 
              loss = 'binary_crossentropy', 
              metrics=['accuracy'])
model.fit(x = inputTF, y = outputTF, batch_size= 160, epochs = 50, verbose=2, validation_split = 0.2, shuffle = True)

#fitting data to our model
```

```python
def prediction(data,verbose = 0):
    predict = pd.DataFrame(data, index=[0])
    predictTF = tf.convert_to_tensor(predict)
    predictions = model.predict(predictTF, verbose = 0)
    p = np.argmax(predictions, axis=-1)
    
    if verbose == 0:
        return outputdata[p[0]]
    elif verbose == 1:
        return predictions
    elif verbose == 2:
        print(f"{round(predictions[0][0]*100,1)}% chance of Normal, {round(predictions[0][1]*100,1)}% chance of Rain, {round(predictions[0][2]*100,1)}% chance of Thunderstorm, {round(predictions[0][3]*100,1)}% chance of Fog, {round(predictions[0][4]*100,1)}% chance of Snow")

#converting model output to readable prediction
```
